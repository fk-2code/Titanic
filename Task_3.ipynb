```python
df=pd.read_csv("/content/Titanic-Dataset.csv")
df.head()
```

```python
df.info()
df.describe()
```

##  Data Cleaning

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.drop(columns=["Cabin"])

df.drop_duplicates()

df.isnull().sum()
```

## Data Analysis

```python
print(df.groupby("Sex")["Survived"].mean())
print(df.groupby("Pclass")["Survived"].mean())
print(df.groupby("Pclass")["Age"].mean())
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 60, 100], labels=['Child', 'Teen', 'Adult', 'Senior'])
print(df.groupby('AgeGroup', observed=False)['Survived'].mean())
```

## Filtering

```python
females_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]

females_survived.head()
```

```python
children_survived = df[(df['Age'] < 18) & (df['Survived'] == 1)]

children_survived.head()
```

```python
first_class_survived = df[(df['Pclass'] == 1) & (df["Survived"] == 1)]

first_class_survived.head()

```
